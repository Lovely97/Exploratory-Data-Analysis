# Data cleaning 


## Removing Null values

```python
# BEFORE REMOVING NULL VALUES OF AGE WE FIRST FIND THE RELATION BETWEEN PASSENGER CLASS ANA AGE,WHAT IS THE AVERAGE AGE OF EACH 
#PASSENGER CLASS 
```

```python
plt.figure(figsize=(12,7))
sns.boxplot(x='Pclass',y='Age',data=train)
```

```python
# 50 percentile is the average value 
```

```python
# NaN VALUES WILL BE REPLACED BY AVERAGE VALUES  OF AGE  OF RESPECTIVE PASSENGERS CLASSES 
# WE CREATE A FUNCTION FOR THIS 
def input_age(cols):
    Age=cols[0]
    Pclass=cols[1]
    
    if pd.isnull(Age):
        
        if Pclass == 1:
            return 37
        
        elif Pclass == 2:
            return 29
        
        else:
            return 24 
        
    else:
            return Age
            
```

```python
# APPLYING THE FUNCTION
train["Age"]= train[['Age','Pclass']].apply(input_age,axis=1)
```

```python
# CHECKING HEATMAP AGAIN  
sns.heatmap(train.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
#NOW ONLY CABIN CONTAINS NAN VALUES ,SINCE THEY ARE SO MANY NAN VALUES WE HAVE TO USE FEATURE 
#ENGINEERING WHICH INVOLVES A LOT OF ANALYSIS SO AT THIS INSTANT OF TIME WE JUST DROP THE COLUMN CABIN
```

```python
def square(n):
    for i in range(n):
        return i**2
```

```python
# TO MAKE ALL ELEMENTS ACCESED WE NEED TO WRITE YIELD IN PLACE OF RETURN ,YIELD IS A GENERATOR 
# GENERATOR IS USED TO MAKE THE ITERATOR
def square(n):
    for i in range(n):
        yield i**2
```

```python
square(2)
```

```python
 for i in square(2):
        print(i)
```
